```python
def missing_values_table(df):
        mis_val = df.isnull().sum()
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Пропущенные значения', 1 : '% от всех записей'})
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% от всех записей', ascending=False).round(1)
        print ("Датафрейм имеет " + str(df.shape[1]) + " колонок.\n"      
            "" + str(mis_val_table_ren_columns.shape[0]) +
              " колонок имеют пропущенные значения.")
        return mis_val_table_ren_columns
```

```python
missing_values_table(data)
```

```python
first_quartile = data['tm'].describe()['25%']
first_quartile
```

```python
third_quartile = data['tm'].describe()['75%']
third_quartile
```

```python
iqr = third_quartile - first_quartile
iqr
```

```python
data = data[(data['tm'] > (first_quartile - 3 * iqr)) &
            (data['tm'] < (third_quartile + 3 * iqr))]
data
```

```python
plt.hist(data['tm'], bins = 20, edgecolor = 'k');
plt.xlabel('Tm'); plt.ylabel('Number of proteins'); 
plt.title('Tm Distribution');
```

## Cipher amino acids to number

```python
amino='FLIMVSPTAYHQNKDECWRG'
for a in amino:
    data[a] = data.apply(lambda x: (x['protein_sequence'].count(a) / len(x['protein_sequence'])), axis=1)
```

```python
data.head()
```

```python
plt.subplots(figsize=(20,20))
sns.heatmap(data.corr(numeric_only=True), fmt=".2f", annot=True, square=True)
plt.show()
```

```python
data = data.drop(columns=['protein_sequence'])
```

## Split data for train and test

```python
import numpy as np
```

```python
X = np.array(data.drop(columns=['tm']))
X
```

```python
y = data['tm'].values
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Try some models

### KNN

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
np.linspace(1, 100, 20)
```

```python
knn_grid = {'n_neighbors': np.array(np.linspace(1, 40, 20), dtype='int')}
```

```python
knn_gs = GridSearchCV(knn, knn_grid, cv=5, n_jobs=6)
```

```python
y_train_knn = y_train.astype('int')
```

```python
knn_gs.fit(X_train, y_train_knn)
```

```python
pd.DataFrame(knn_gs.cv_results_)
```

```python
def grid_plot(x, y, x_label, title, y_label='cross_val'):
    plt.grid(True)
    plt.plot(x, y, 'go-')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
```

```python
grid_plot(knn_grid['n_neighbors'], knn_gs.cv_results_['mean_test_score'], 'n_neighbors', 'KNeighborsClassifier')
```

#### Low score, not a good model

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1)
```

```python
rf_grid = {'n_estimators': np.array(np.linspace(5, 100, 5), dtype='int')}
rf_gs = GridSearchCV(rf, rf_grid, cv=5, n_jobs=6, scoring="accuracy")
rf_gs.fit(X_train, y_train_knn)
```

```python
pd.DataFrame(rf_gs.cv_results_)
```

```python
grid_plot(rf_grid['n_estimators'], rf_gs.cv_results_['mean_test_score'], 'n_estimators', 'RandomForest')
```

### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
np.linspace(50, 500, 4)
```

```python
rfr = RandomForestRegressor()
rfr_grid = {'n_estimators': np.array(np.linspace(50, 500, 4), dtype='int'), 
            'max_depth': [5, 10, 15, None]}
rfr_gs = GridSearchCV(rfr, rfr_grid, cv=5, n_jobs=6)
rfr_gs.fit(X_train, y_train)
```

```python
pd.DataFrame(rfr_gs.cv_results_)
```

```python
rfr_gs.best_params_
```

```python
rfr_best = RandomForestRegressor(n_estimators=350, max_depth=None)
rfr_best.fit(X_train, y_train)
```

```python
y_train_pred_rfr = rfr_best.predict(X_train).round()
y_test_pred_rfr = rfr_best.predict(X_test).round()
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test, y_test_pred_rfr)
```
